```python
A = [-4,3,2,0,7,8,9,5,6]

import heapq #By default, heapq in Python only supports a min-heap. That means heap[0] is always the smallest element.
heapq.heapify(A)

A
```

```python
#Heap Push (Insert Element)
#Time : O(log n) 

minn = heapq.heappush(A,11)
A
```

```python
#Heap Pop (Extract min)
#Time : O(log n)

minn = heapq.heappop(A)

A
```
