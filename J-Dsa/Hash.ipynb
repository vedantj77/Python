```python
#hashset (mutable)

s = set()

s.add(1)
s.add(2)
s.add(3)

print(s)


if 1 in s :
    print(True)

s.remove(3)
print(s)

for i in s:
    print(i)
```

```python
from collections import Counter
string = "aaabbbbbbbbbbbssssssssshhhhhhhhhhhhhheeeeeeeeeeeeeee"
sett = set(string)   #string is converted in a set and stored in sett
print(sett)  #to print unique values
Counterr= Counter(string)
print(Counterr)
```

```python
#Hashmaps - Dictionaries stores in key : value

d = {'navin':'apple','soham':2 ,'vedant':3,} 
print(d)

#add key : value pair in d
d['arsh'] = 9
print(d)

#check for presence of key in dictionary : O(1)
if 'navin' in d:
    print(True)


#check the value coressponding to a key in the dictionary : O(1)
print(d['navin'])

#Loop over the key:value pairs of the dictionary : O(n)
for key,val in d.items():
    print(f'key-{key}: val-{val}')
```
